```python

# surface plot for 2d objective function
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LightSource
from matplotlib import cbook, cm


        
# objective function
def objective(x, y):
 return (x - 10)**2.0 +  y**2.0 + 5
 
# define range for input
x_min, x_max = -5.0, 15.0
y_min, y_max = -10.0, 10.0
# sample input range uniformly at 0.1 increments
xaxis = np.arange(x_min, x_max, 0.1)
yaxis = np.arange(y_min, y_max, 0.1)

r_opt = np.linspace(15, 0.0, 15)
theta_opt = np.linspace(-np.pi / 2 , np.pi / 2, 15)
x_opt = (r_opt) * np.sin(theta_opt) + 10
y_opt = r_opt * np.cos(theta_opt)
z_opt = objective(x_opt, y_opt)


# create a mesh from the axis
x, y = np.meshgrid(xaxis, yaxis)
# compute targets
z = objective(x, y)
# create a surface plot with the jet color scheme
figure = plt.figure()
axis = figure.add_subplot(projection='3d', computed_zorder=False)



ls = LightSource(azdeg=0, altdeg=45)
# To use a custom hillshading mode, override the built-in shading and pass
# in the rgb colors of the shaded surface calculated from "shade".
rgb = ls.shade(z, cmap=cm.YlOrBr, vert_exag=0.1, blend_mode='soft')
axis.plot_surface(x, y, z,  #lw=0.5, rstride=50, cstride=50, 
                  # alpha=0.2, 
                  edgecolor='none',
                  facecolors=rgb,
                  antialiased=False,
                 zorder=-1000)
axis.plot(x_opt, y_opt, z_opt, 'b', linewidth=2, zorder=10)
axis.scatter(x_opt[-1], y_opt[-1], z_opt[-1], marker='o', zorder=1000, s=50, c='r', zdir='z', alpha=1.)
# axis.contourf(x, y, results, zdir='z', offset=0, cmap='coolwarm')
axis.set(xlabel=r'$z_1$', ylabel=r'$z_2$', zlabel=r'$E(z_1,z_2)$', xticklabels=[], yticklabels=[], zticklabels=[])
axis.grid(False)
# axis.view_init(40, 290)
axis.set_box_aspect(aspect=None, zoom=0.8)


# show the plot
plt.savefig('img.png', bbox_inches='tight');

plt.show()
```
