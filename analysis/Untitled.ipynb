```python
import matplotlib.pyplot as plt
import numpy as np


def smooth_square_sin(i, N=33, eps=0.5):
    a = (i/N) * np.pi
    return np.sin(a) / np.sqrt(np.sin(a)**2 + eps)
```

```python
t = np.arange(100)
plt.plot(t, (1 + smooth_square_sin(t))/2 + 1e-3)
plt.plot(t, (1 - smooth_square_sin(t))/2 + 1e-3)
```
